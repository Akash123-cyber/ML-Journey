# 🧩 1. Introduction to Regression Analysis

### 📘 Definition

- **Regression Analysis is a supervised machine learning technique used to find the relationship between dependent (target) and independent (predictor) variables.**

### 💡 In simple terms:

**Regression** helps us predict a **continuous numeric output** (like price, salary, temperature, etc.) based on input variables.

### 🎯 Example

| Experience (Years) | Salary ($) |
| ------------------ | ---------- |
| 1                  | 40,000     |
| 2                  | 45,000     |
| 3                  | 60,000     |
| 4                  | 80,000     |


Here,

- Independent variable (X): Experience

- Dependent variable (Y): Salary

**Regression helps us learn a function Y = f(X) that best fits the data.**

# 🔍 Comparison: Linear vs Non-Linear Regression Analysis



| Feature | **Linear Regression** | **Non-Linear Regression** |
|----------|------------------------|----------------------------|
| **Relationship Between Variables** | Assumes a straight-line (linear) relationship between dependent and independent variables. | Assumes a curved or complex relationship that cannot be represented by a straight line. |
| **Equation Form** |  $$ Y = \beta_0 + \beta_1 X + \epsilon $$ |$$ Y = \beta_0 + \beta_1 X + \beta_2 X^2 + \dots + \epsilon $$ or even exponential/logarithmic forms like $$Y = \beta_0 e^{\beta_1 X}$$|
| **Model Type** | First-degree polynomial (straight line). | Higher-degree polynomial or non-polynomial (e.g., exponential, logarithmic, power, logistic). |
| **Graph Shape** | Straight line. | Curve (can be quadratic, exponential, logarithmic, etc.). |
| **Interpretability** | Simple and easy to interpret. | More complex; interpretation depends on the model form. |
| **Computation** | Easier and computationally less expensive. | More computationally intensive; requires iterative fitting methods. |
| **When to Use** | When data shows a roughly linear pattern. | When data shows curvature or non-linear trends. |
| **Example Use Case** | Predicting salary based on experience, house price vs area. | Predicting population growth, enzyme reaction rates, disease spread, or sales over time. |
| **Mathematical Solution** | Can be solved using Ordinary Least Squares (OLS). | Requires numerical optimization techniques (e.g., Gradient Descent, Levenberg–Marquardt). |
| **Performance on Curved Data** | Poor fit (underfits data). | Better fit for data with non-linear patterns. |





# ⚖️ Classification of Regression Analysis Algorithms
### Based on the Nature of Relationship — Linear vs Non-Linear

Regression algorithms are broadly classified into two main categories depending on whether the relationship between the dependent and independent variables is **linear** or **non-linear**.

---

### 🔹 1. Linear Regression Algorithms

Linear Regression models assume a **straight-line relationship** between the independent variable(s) (X) and the dependent variable (Y).  
They predict outputs using a linear combination of input features.

$$
Y = \beta_0 + \beta_1 X_1 + \beta_2 X_2 + \dots + \beta_n X_n + \epsilon
$$

| Algorithm | Description | Characteristics | Example Use Case |
|------------|--------------|------------------|------------------|
| **Simple Linear Regression** | Predicts Y using one independent variable X. | Straight-line fit, single feature. | Predicting salary based on experience. |
| **Multiple Linear Regression** | Uses multiple independent variables to predict Y. | Multi-dimensional linear model. | Predicting house price using area, rooms, and location. |
| **Ridge Regression (L2 Regularization)** | Adds L2 penalty term to reduce overfitting. | Shrinks coefficients but keeps all features. | When data is multicollinear. |
| **Lasso Regression (L1 Regularization)** | Adds L1 penalty term to induce sparsity. | Eliminates less important features. | Feature selection in high-dimensional data. |
| **ElasticNet Regression** | Combination of Ridge and Lasso penalties. | Balances coefficient shrinkage and feature selection. | Large datasets with correlated variables. |
| **Stepwise Regression** | Adds/removes predictors based on statistical significance. | Automated variable selection technique. | When many potential predictors exist. |

---

### 🔹 2. Non-Linear Regression Algorithms

Non-Linear Regression models capture **curved or complex relationships** between X and Y.  
They don’t have to follow a straight-line pattern; instead, they model more flexible, often non-parametric, relationships.

| Algorithm | Description | Relationship Type | Example Use Case |
|------------|--------------|-------------------|------------------|
| **Polynomial Regression** | Models Y as a polynomial function of X. | Quadratic, cubic, or higher-order relationships. | Predicting growth trends, curved patterns. |
| **Exponential Regression** | Models Y = a × e^(bX). | Exponential growth or decay. | Population growth, radioactive decay. |
| **Logarithmic Regression** | Models Y = a + b × ln(X). | Logarithmic relationships. | Diminishing returns, learning curves. |
| **Power Regression** | Models Y = a × X^b. | Power-law relationships. | Physics and biology growth processes. |
| **Support Vector Regression (SVR)** | Uses kernel trick (polynomial/RBF) to model non-linear relationships. | Depends on kernel choice. | Predicting temperature or energy consumption. |
| **Decision Tree Regression** | Splits data into regions and fits constant values. | Piecewise non-linear. | Predicting prices, demand forecasting. |
| **Random Forest Regression** | Ensemble of multiple trees. | Non-linear and robust. | Predicting sales, complex numeric targets. |
| **Gradient Boosting Regression (XGBoost, LightGBM)** | Sequential ensemble of weak learners (trees). | Complex, flexible non-linear relationships. | Advanced predictive analytics. |
| **K-Nearest Neighbors (KNN) Regression** | Predicts Y by averaging K nearest neighbors. | Instance-based non-linear. | Simple pattern recognition tasks. |
| **Neural Network Regression** | Uses neurons and activation functions to learn non-linear mappings. | Highly flexible, complex non-linear. | Predicting stock prices, image-based numeric outputs. |

---

# ⚙️ 2. Understanding Linear Regression

**Linear Regression** is the simplest form of regression.
It assumes a **linear relationship** between input (X) and output (Y):

#### $$ Y = \beta_0 + \beta_1 X + \epsilon $$


where:

**𝑌 → dependent variable (what we predict)**

**𝑋 → independent variable**

**𝛽0 → intercept (value of Y when X=0)**

**𝛽1 → slope (change in Y per unit change in X)**

**𝜖 → error term (difference between predicted and actual value)**

<hr>



# 🧮 3. Mathematical Intuition Behind Linear Regression

The goal is to find the **best-fitting** line through the data points.

That means finding **𝛽0** and **𝛽1** that **minimize the Sum of Squared Errors (SSE)**:

### 🎯 Sum of Squared Errors (SSE)

$$
SSE = \sum_{i=1}^{n} (y_i - (\beta_0 + \beta_1 x_i))^2
$$

### ⚙️ Formula for Coefficients (Ordinary Least Squares)

$$
\beta_1 = \frac{\sum (x_i - \bar{x})(y_i - \bar{y})}{\sum (x_i - \bar{x})^2}
$$

$$
\beta_0 = \bar{y} - \beta_1 \bar{x}
$$

<hr>



# 📜 4. Assumptions of Linear Regression

- Linear Regression works well only if the following assumptions hold true:


  
| No | Assumption               | Meaning                                        |
| -- | ------------------------ | ---------------------------------------------- |
| 1  | **Linearity**            | Relationship between X and Y is linear         |
| 2  | **Independence**         | Observations are independent                   |
| 3  | **Homoscedasticity**     | Constant variance of residuals                 |
| 4  | **Normality of Errors**  | Residuals are normally distributed             |
| 5  | **No Multicollinearity** | Independent variables aren’t highly correlated |

# 📊 5. Evaluating Model Performance
To check how good your regression model is:

| Metric                        | Formula                                | Ideal Value     | 
| ----------------------------- | -------------------------------------- | --------------- | 
| **R² (R-squared)**            | $$( 1 - \frac{SSR}{SST} )$$                | Closer to 1     | 
| **RMSE (Root MSE)**           | $$( \sqrt{MSE} )$$                         | Lower is better | 
| **MSE (Mean Squared Error)**  | $$( \frac{1}{n}\sum(y_i - \hat{y}_i)^2 )$$ | Lower is better | 
| **MAE (Mean Absolute Error)** | $$( \frac{1}{n}\sum\| y_i - \hat{y}_i \| )$$ | Lower is better |

# **🧪 9. Checking Model Assumptions (Residual Analysis)**

- ##### After fitting a **Linear Regression model**, we cannot just stop at the equation — we must **verify that our model actually fits the data well** and that it **follows the assumptions** of linear regression.

- #####  **Residual analysis** is the **process of examining the residuals** (the **differences between actual and predicted values**) to ensure that the model assumptions hold true.

<hr>

## ⚙️ What are Residuals?

- ##### **A residual is simply the error for each data point**:

### 📉 Residual Formula

$$
e_i = y_i - \hat{y_i}
$$

where  

- $y_i$ = actual value  
- $\hat{y_i}$ = predicted value  
- $e_i$ = residual (error)  

Residuals help us understand what the model **could not explain**.

<hr>

### 🧩 Key Assumptions of Linear Regression

- Let’s explore each one and what it means 👇

####  1️⃣ Linearity

- The relationship between **independent (X)** and **dependent (Y)** variables should be linear.

- In other words, **Y should change at a constant rate with respect to X**.

**How to check:**

- **Plot residuals vs. predicted values** — if they are **randomly scattered around zero (no pattern), linearity is satisfied**.
- If you **see curves, waves, or systematic patterns, it indicates non-linearity**.

---

#### 2️⃣ Homoscedasticity (Equal Variance)

- The residuals should have **constant variance** across all levels of predicted values.  
- If variance increases or decreases, it means **heteroscedasticity**, which can bias results.

**How to check:**  
Plot residuals vs. predicted values.  
✅ Homoscedasticity → random scatter  
❌ Heteroscedasticity → cone or funnel shape  

---

#### 3️⃣ Independence of Errors

- The residuals should be **independent** of each other.  
- If errors are correlated (like in time series), it violates this assumption.

**How to check:**  
Use the **Durbin-Watson test** — a value around **2** indicates independence.  

---

#### 4️⃣ Normality of Residuals

- The residuals should follow a **normal distribution** (bell-shaped curve).  
- This is important for accurate confidence intervals and hypothesis testing.

**How to check:**  
- Use a **Histogram** or **Q-Q Plot** of residuals.  
- Or perform the **Shapiro–Wilk test**.

---

#### 5️⃣ No Multicollinearity (for multiple regression)

- The independent variables should not be **highly correlated** with each other.

**How to check:**  
Use **Variance Inflation Factor (VIF)** —  
✅ $VIF < 5$ (good)  
❌ $VIF > 10$ (strong multicollinearity problem)


## ➡️ **Now Implementation**

### **1. Residual plot Linearity And Homoscedasticity Check ✅**

##### ➡️ first calculate the residuals

````python
residuals = y - y_pred
````

````python
#1. Residual plot

plt.scatter(y_pred, residuals)
plt.axhline(y = 0, color = 'r', linestyle = '--')
plt.xlabel("Predicted Values")
plt.ylabel("Residuals")
plt.title("Residuals vs Fitted")
plt.show()
````

### 🌟🌟🌟 NOTE FOR ABOVE PLOT:

#### **Explanation of `.axhline()`**

The `.axhline()` function is a method in Matplotlib (often imported as `plt`) used to draw a straight horizontal line across the entire plot.

The arguments used are:

- `y=0`: Specifies the Y-coordinate where the line should be drawn. In a residuals plot, this is the most important position,**representing zero residual (perfect prediction)**.
- `color='r'`: Sets the color of the line to **red**.
- `linestyle='--'`: Sets the line to a **dashed** style.

**Interpretation**: For a good regression model, the residuals should be randomly scattered around this red $Y=0$ line which this plot shows meaning **the data/plot follows the assumptions of linearity and homoscedasticity**. If the dots form a pattern (like a curve or a cone shape), it indicates problems with the model, such as non-linearity or heteroscedasticity.

### **2. Histogram of Residuals : Normality Check of Residuals ✅**

````python
sns.histplot(
    residuals,
    stat = "count",
    # stat = "density",
    kde = True,
    kde_kws = dict(cut = 3),
    edgecolor = (1,1,1,0),
    alpha = 0.4,
    bins = 50
    )
plt.title("Distribution of Residuals")
plt.show()
````

### 🌟🌟🌟 NOTE FOR ABOVE HIST PLOT

**Interpretation of this Hist Plot:**
In an ideal regression model, the residuals should be **normally distributed with a mean of zero**. Let's see how this plot aligns with that:

- **Symmetry and Central Tendency**: The histogram and KDE curve are roughly symmetric and centered around **zero** (the peak of the curve is near 0 on the x-axis). This is a good sign, as it suggests the model is not systematically over- or under-predicting.

- **Shape (Bell Curve)**: The **overall shape resembles a bell curve**, indicating that the residuals are approximately normally distributed. Most residuals are clustered around zero, and fewer residuals are found as you move further away from zero. **This is a key assumption for many linear regression models and suggests that the errors are random and not biased.**

- **Spread**: The residuals are spread out, but the majority are concentrated between roughly -0.5 and 0.5, with some outliers extending to -1.0 and 1.0. The density drops off at the extremes.

**In summary, this plot suggests a relatively good fit for a regression model:**

- The residuals are **centered around zero**.

- The residuals appear to be **approximately normally distributed**.

**This indicates that the model's errors are random and well-behaved, which is a desirable characteristic for a robust regression model.**


### **3. Q-Q Plot (Quantile - Quantile Plot) Normality Check ✅**

````python
import statsmodels.api as sm

sm.qqplot(residuals, line = '45', fit = True)
plt.title("Q-Q Plot of Residuals")
plt.show()
````

### 🌟🌟🌟 **Interpretation of the Plot**

#### 1️⃣ **What the plot shows**

- **X-axis (Theoretical Quantiles)**: Expected quantiles if the residuals were perfectly normally distributed.

- **Y-axis (Sample Quantiles)**: Actual quantiles of your regression residuals.

- **Red line**: Represents the ideal normal distribution. If residuals are perfectly normal, all points lie exactly on this line.

#### 2️⃣ **How to interpret it**

- **Points closely following the red line**: Residuals are approximately normally distributed.

- **Points deviating from the line at the ends (tails)**: Indicates some skewness or heavy/light tails compared to normal distribution.

- **S-shaped curve**: Suggests skewness in residuals.

- **Upward/downward curves at ends**: Suggests heavier or lighter tails than normal.

####  3️⃣ Obs**ervations for your plot**

- Most points **closely follow the red line**, **indicating** that residuals are **roughly normally distributed**.

- **Minor deviations** at the extremes (very high or low residuals) are visible, but these **are often acceptable in practice**.

✅ **Conclusion**: **The normality assumption of residuals in our simple linear regression seems reasonably satisfied. This supports the validity of confidence intervals, hypothesis tests, and other inference based on this regression.**

### ➡️ Us**e Cases for Q-Q Plots:**

Q-Q plots are incredibly useful in various statistical analyses, particularly in regression:

1. **Checking Normality Assumption (Most Common)**:

   - **Regression**: Many linear regression models assume that the **residuals are normally distributed**. A Q-Q plot is the best visual tool to check this assumption. If residuals are not normal, it can affect the validity of confidence intervals and hypothesis tests.

   - **Other Statistical Tests**: Many parametric tests (e.g., t-tests, ANOVA) also assume normality of data or residuals.

2. **Comparing Distributions**: You can use Q-Q plots to compare any two distributions to see if they are similar in shape, scale, and location.

3. **Detecting Skewness and Kurtosis**: The way points deviate from the line can tell you about the skewness (asymmetry) or kurtosis (peakedness/tail heaviness) of your data compared to the theoretical distribution.

4. **Identifying Outliers**: Points that fall far away from the straight line at the ends of the plot can indicate outliers in your data.

### 🚦Interpreting the Results

| Test Type              | What You Expect              | If Not Met                                |
| ---------------------- | ---------------------------- | ----------------------------------------- |
| Residuals vs Fitted    | Random scatter around zero   | Try polynomial or nonlinear model         |
| Histogram of residuals | Bell-shaped (normal)         | Try transformation or different model     |
| Q-Q Plot               | Points close to the 45° line | Skewed or non-normal errors               |
| Durbin-Watson          | Around 2                     | Errors correlated (common in time series) |
| VIF                    | < 5                          | Strong correlation among X variables      |
