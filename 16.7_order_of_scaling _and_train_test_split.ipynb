### Why Itâ€™s Wrong
- The scaler learns statistics (mean, std, min, max, etc.) from **ALL data**, including the test set.
- Your model indirectly â€œknowsâ€ something about the test data before training.

> This is called **data leakage**.

It produces artificial performance improvement and poor real-world generalization.

## ðŸ’¡ Important Rule

> Any preprocessing that learns parameters must be fit only on training data.

**Examples:**

- StandardScaler
- MinMaxScaler
- RobustScaler
- PCA
- Feature selection
- Normalization
