----
### **1. Generally we have two types of data Numerical and Categorical(String type) data in Machine Learning it is known as Object Data**

### **2. To handle Object type of data we generally try to do it with mode: we try to fill the most frequent value for the missing object data type**

### **3. Other then mode method we also uses the forward filling and backward filling method for Object data type missing value**

***In Backward Filling : the data next to the missing value in downward fashion is filled backward basically inplace of missing value***

***In Forward Filling : the data previous to the missing value in downward fashion inside a column is filled forward basically in place of missing value***

## NOTE
- ### These methods are used for both numerical and categorical data
    - #### `--->` For Numbers and,
    - #### `--->` For String Data too

### The Reason Why Warning Is Appearing For This Code :
```python
df[i].fillna(df[i].mode()[0], inplace=True)
```
### And Not For this Code :

```python
df["Gender"].fillna(df["Gender"].mode()[0], inplace=True)
```
----
### IS BECAUSE: of how Pandas handles column access and the difference between explicit and dynamic indexing.

#### Key Difference Between the Two Examples

#####  1. `df["Gender"]` (Explicit Columns Access):
- When you use explicit column access like `df["Gender"]`, Pandas accesses the column directly from the original DataFrame. This means it is not creating a temporary copy of the column.

- As a result, when you apply `.fillna(..., inplace=True)`, the operation works directly on the original DataFrame, and no warning is raised.

##### 2. `df[i]` (Dynamic Column Access):
- When you use dynamic column access like `df[i]`, Pandas treats it as a "view" or a "temporary copy" of the column. This happens because `i` is a variable (e.g., `i = "Gender"`) and Pandas cannot always guarantee whether the operation is happening on the original DataFrame or a copy.




- In such cases, using `.fillna(..., inplace=True)` raises a warning because Pandas cannot ensure that changes will affect the original DataFrame.

##### Why Is This a Problem for Dynamic Indexing?

- Dynamic indexing `(df[i])` creates a temporary object in memory. If you use `inplace=True` on this temporary object, Pandas cannot confidently apply the operation to the original DataFrame, hence the warning.

- Explicit indexing `(df["Gender"])`directly references the column in the original DataFrame, so no temporary object is created, and the operation works as intended.

##### Solution to Avoid the Warning

- If you are using dynamic indexing (i.e., `i` is a variable containing the column name), avoid `inplace=True` and assign the result back to the column:

```python
df[i] = df[i].fillna(df[i].mode()[0])
```
##### Why This Fix Works?
- `df[i] =`: You explicitly assign the result back to the column `i` in the DataFrame, ensuring that the changes are applied directly to the original ``df``.

- This avoids any confusion about whether you're working on a copy or the original DataFrame.

## NOTE FOR ABOVE CODE :: 

- ### This inplace method will never work because the intermediate object on which we are setting values always behaves as a copy.

- ### For example, when doing 
    ```python 
    df[col].method(value, inplace=True)
    ```
- ### try using 
    ```python
    df.method({col: value}, inplace=True)
    ``` 
- ### or 
    ```python
    df[col] = df[col].method(value)
    ``` 
- ### instead, to perform the operation inplace on the original object.



````python
# CORRECT WAY
null_counts = df.select_dtypes(include = "object").isnull().sum()

column_names = null_counts[null_counts > 0].index.tolist()

column_names
````

````python
# Correct way

for col in column_names:
    df.fillna({col:df[col].mode()[0]}, inplace = True)
````
