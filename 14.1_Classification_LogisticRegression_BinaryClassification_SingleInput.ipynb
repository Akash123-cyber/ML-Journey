- **Multilabel Classification :** multiple outputs for each instance(rare)

---

## üß© Classification of Classification Algorithms Based on **Linearity**

Classification algorithms can be broadly categorized into **Linear** and **Non-Linear** models based on the **shape of their decision boundary** ‚Äî that is, how they separate different classes in the feature space.

---

### üîπ 1. **Linear Classification Algorithms**

These algorithms assume that the classes can be separated by a **straight line (in 2D)** or a **hyperplane (in higher dimensions)**.

#### üß† Key Concept

They try to find parameters (weights) such that the equation:

$$
w_1x_1 + w_2x_2 + \dots + w_nx_n + b = 0
$$

best separates the classes.

#### ‚úÖ Examples of Linear Classifiers

| Algorithm | Description |
|:--|:--|
| **Logistic Regression** | Uses a linear decision boundary and the sigmoid function to model probabilities. |
| **Linear SVM (Support Vector Machine)** | Finds the hyperplane that maximizes the margin between classes. |
| **Perceptron** | A single-layer neural network that separates linearly separable data. |
| **Linear Discriminant Analysis (LDA)** | Projects data to a lower-dimensional space where classes are most separable using linear combinations. |

#### üí° Characteristics
- Fast and simple to implement.  
- Works well when data is **linearly separable**.  
- Struggles with complex, curved decision boundaries.  
- Easy to interpret (weights show feature importance).

---

### üî∏ 2. **Non-Linear Classification Algorithms**

These algorithms can model **complex, curved, or irregular decision boundaries**.  
They use **kernel tricks**, **tree structures**, or **layered transformations** to handle **nonlinear separability**.

#### ‚úÖ Examples of Non-Linear Classifiers

| Algorithm | Description |
|:--|:--|
| **Kernel SVM (e.g., RBF Kernel)** | Uses the kernel trick to map input data into higher dimensions where it becomes linearly separable. |
| **K-Nearest Neighbors (KNN)** | Makes predictions based on the majority class among nearby samples ‚Äî no fixed decision boundary. |
| **Decision Tree** | Splits data recursively based on features ‚Äî inherently non-linear. |
| **Random Forest** | Ensemble of multiple Decision Trees ‚Äî combines non-linear boundaries. |
| **Na√Øve Bayes** | Though it assumes conditional independence, its boundary is often non-linear. |
| **Neural Networks (Deep Learning)** | Uses multiple layers and activation functions to model highly complex, non-linear relationships. |

#### üí° Characteristics
- Can capture complex, real-world relationships.  
- Often require more data and computational power.  
- Risk of overfitting if not properly regularized.  
- Harder to interpret compared to linear models.

# ‚öôÔ∏è 2. Logistic Regression ‚Äî Concept & Theory

Despite the name, **Logistic Regression is used for Classification**, not Regression.

### üß† Key Idea

We model the **probability** that an observation belongs to a class.

$$
P(y=1|x) = \sigma(z) = \frac{1}{1 + e^{-z}}
$$

Where

$$
z = \beta_0 + \beta_1x_1 + \beta_2x_2 + \dots + \beta_nx_n
$$

The **sigmoid (logistic) function** squashes any real number $z$ between **0 and 1**, which can then be interpreted as probability.

If  

- $P(y=1|x) \ge 0.5 \Rightarrow y=1$  
- $P(y=1|x) < 0.5 \Rightarrow y=0$


---

## <center>üèÜ Types Of Logisitc Regression</center>
**"On the basis of the categories, Logistic Regression can be calssified into three types"**

1. **Binomial:** In binomial Logistic Regression, there can be only two possible types of the dependent variables, such as 0 or 1, Pass or Fail, etc

2. **Multinomial:** In Multinomial Logistic Regression, there can be 3 or more possible unordered types of the dependent variables such as "cat","dogs", or "sheep".

3. **Ordinal:** In ordinal Logistic Regression there can be 3 or more possible ordered types dependent vairables such as "low", "medium", or "high".

# üßÆ 3. Mathematics Behind Logistic Regression

### üîπ Step 1: Linear Combination :
üåüüåüüåü **For this we use the Linear Regression for the combination of different variables and then we feed the output of the Linear Regression to the `sigmoid function` which squashes any number between 0 and 1 so we can asses it as probability**

$$
z = \beta_0 + \beta_1x_1 + \beta_2x_2 + \dots + \beta_nx_n
$$

### üîπ Step 2: Apply Sigmoid

$$
\sigma(z) = \frac{1}{1 + e^{-z}}
$$

*Output lies between 0 and 1.*

### üîπ Step 3: Interpretation

$$
P(y=1|x) = \sigma(z)
$$

$$
P(y=0|x) = 1 - \sigma(z)
$$

---

# üí° 4. Cost Function ‚Äî Log Loss

Since this is **not a linear problem**, we **cannot use MSE** as a cost function (it gives a non-convex curve).

We use **Log Loss** (Binary Cross-Entropy Loss):

$$
J(\beta) = -\frac{1}{m} \sum_{i=1}^{m} \left[ y_i \log(\hat{y_i}) + (1 - y_i)\log(1 - \hat{y_i}) \right]
$$

Where:

- $y_i$: actual label (0 or 1)  
- $\hat{y_i}$: predicted probability  
- $m$: number of training samples  

The function is **convex** ‚Äî meaning gradient descent can find a global minimum.

---

# üß∞ 5. Model Training ‚Äî Gradient Descent

We optimize parameters $\beta$ using **Gradient Descent**:

$$
\beta_j = \beta_j - \alpha \frac{\partial J(\beta)}{\partial \beta_j}
$$

Where:

- $\alpha$: learning rate  
- The derivative $\frac{\partial J(\beta)}{\partial \beta_j}$ tells how much to change $\beta$ to reduce cost.

---


# üß† 6. Model Evaluation Metrics

For **classification**, accuracy alone isn‚Äôt enough ‚Äî especially for **imbalanced data.**

**Common Metrics:**

| Metric               | Formula                                         | Description                                                                |
| :------------------- | :---------------------------------------------- | :------------------------------------------------------------------------- |
| Accuracy             | (TP + TN)/(TP + TN + FP + FN)                   | Overall correctness                                                        |
| Precision            | TP / (TP + FP)                                  | Out of predicted positives, how many are correct                           |
| Recall (Sensitivity) | TP / (TP + FN)                                  | Out of actual positives, how many were found                               |
| F1 Score             | 2 √ó (Precision √ó Recall) / (Precision + Recall) | Harmonic mean                                                              |
| ROC-AUC              | Area under ROC Curve                            | Probability that model ranks a random positive higher than random negative |

# ‚û°Ô∏è **Another Example Of `Binomial Logistic Regression`**

```python
#step 1: IMport Libraries
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, roc_auc_score
```

```python
#step2: load dataset

from sklearn.datasets import load_breast_cancer

dataset = load_breast_cancer()
```

```python
print(dataset.DESCR)
```

```python
x = dataset.data
y = dataset.target
```

```python
x = pd.DataFrame(data = x,columns = dataset.feature_names )
```

```python
x.head()
```

```python
# step3: Split dataset into training and testing
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
```

```python
#Step 4: Initialize the model

model = LogisticRegression(max_iter = 1000)
```

```python
#step5: Train the model
model.fit(x_train, y_train)
```

```python
#step6: Make predictions
y_pred = model.predict(x_test)
y_pred_proba = model.predict_proba(x_test)[:,1] #probabilities
```

```python
#step7: Evaluation

print(f"accuracy_score : {accuracy_score(y_test, y_pred)}")
print(f"\nconfusion_matrix :\n{confusion_matrix(y_test, y_pred)}")
print(f"\nClassfication Report:\n{classification_report(y_test, y_pred)}")
print(f"\nROC-AUC Score: {roc_auc_score(y_test, y_pred_proba)}")
```
