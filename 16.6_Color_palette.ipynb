# ðŸŒˆ Color Palette Followed in Matplotlib and Seaborn

### **1. Categorical Palettes (Best for distinct groups)**
These are best for distinct groups (like Setosa vs. Virginica) where no group is "greater" than another.

- **'Set2'**: (My personal favorite) Soft, distinct pastel colors that are easy on the eyes.
- **'tab10'**: The standard "Tableau" colors. Very bold and distinct.
- **'pastel'**: Very light, washed-out colors. Good if you have dark text over bars.
- **'deep'**: The default Seaborn palette (muted but clear).
- **'Paired'**: Good if you have pairs of related categories (light blue, dark blue, light green, dark green).

---

### **2. Sequential Palettes (Best for Continuous Data)**
These are best when the data goes from low to high (like a heatmap or age).

- **'viridis'**: The modern gold standard. High contrast, colorblind-friendly, goes from purple to yellow.
- **'magma'**: Similar to viridis but uses black-red-yellow (looks like lava).
- **'Blues', 'Greens', 'Reds'**: Simple single-color gradients (Note the capital first letter and plural 's').
- **'rocket'**: A Seaborn specific palette (dark red to light orange).
- **'mako'**: A Seaborn specific palette (dark teal to light green).

---

### **3. Diverging Palettes**
Best for data with a center point (like positive vs. negative correlation).

- **'coolwarm'**: The classic Blue-to-Red.
- **'RdBu'**: Red-to-Blue (often better for colorblindness than Red-Green).
- **'Spectral'**: A rainbow variation that diverges.

---

### **Pro Tip: Reversing Colors**
You can reverse any palette by adding `_r` to the end of the name.

Example: `'viridis_r'`, `'Blues_r'`, `'Set2_r'`.

---

### **How to preview them quickly**
If you want to see what a palette looks like without plotting your whole graph, you can run this distinct line in your notebook:

```python
# Shows the 'Set2' palette
sns.color_palette("Set2")

```

When you run `sns.color_palette("Set2")`, it returns a **list of RGB tuples** ðŸ“Œ  
Because it is a **standard Python list**, you can select specific colors using **indexing** (just like any other list).

This is extremely useful when:
- You don't like all colors in a palette ðŸŽ¨
- You want consistent colors across multiple plots
- You want to highlight certain categories

---

## ðŸŒŸ Code to Select Specific Colors from a Palette


````python
import seaborn as sns
import matplotlib.pyplot as plt
````

### 1. Capture the palette in a variable
#### This variable 'set2' is now a list of colors: [color1, color2, color3, ...]

````python
set2 = sns.color_palette("Set2")
set2
````

### 2. Select the specific colors you want by index (0-based)
#### Example: Let's pick the 1st (index 0), 2nd (index 1), and 6th (index 5) color

````python
my_selection = [set2[0], set2[1], set2[5]]
````

### 3. Pass your selection to the palette argument

````python
iris = sns.load_dataset("iris")

sns.countplot(
    x = 'species',
    data = iris,
    hue = 'species',
    palette = my_selection
)

plt.title("Using Specific Colors from Set2")
````

#### **ðŸ” Result Explanation**

By selecting specific indices, you **manually control which colors appear** in the plot.
This gives full flexibility and prevents seaborn from auto-assigning random ones.

Example:

- You can force `yes/no`

- Or multiple classes with consistent color mapping

- Or maintain brand or UI color consistency

### ðŸ§  How to Know Which Index Is Which Color?

Colors are stored **in the order the palette displays them**.  
Indices start from `0` âž¡ï¸ left â†’ right.

---

## **ðŸ§ª Visualizing Qualitative (Categorical) Palettes**

##### These are best when categories are distinct and not ordered
> ####  âž¤ Like Iris species, customer type, sentiment labels, etc.

````python
import seaborn as sns
import matplotlib.pyplot as plt

# These are the best ones for categorical data (like Iris Species)

qualitative_palettes = ['Set1', 'Set2', 'Set3', 'Pastel1', 'Pastel2', 'Paired', 'Accent', 'Dark2', 'tab10']

#creating a plot to show them

fig, axes  = plt.subplots(len(qualitative_palettes), 1, figsize= (6,5))

for i, name in enumerate(qualitative_palettes):
    # Visualize the palette
    sns.heatmap(
        [range(8)], cmap = name, cbar = False, ax = axes[i], xticklabels=False, yticklabels=False
    ),
    
    axes[i].set_ylabel(name, rotation = 0, ha = 'right', fontsize = 12)

plt.tight_layout()
plt.show()
````
