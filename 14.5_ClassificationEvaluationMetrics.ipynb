## <center> **‚û°Ô∏è CONFUSION MATRIX** </center>

> ### **It is the foundation of classification evaluation**.

> #### A confusion matrix is a simple and useful tool/metric for understanding the performance of a classification model.

> #### It helps you evaluate how well your model is doing in categorizing things correctly.

> #### It is also known as the **error matrix**.

> #### The matrix consists of predictions result in a summarized form
> #### **For a binary classifier, the confusion matrix looks like:**

|                     | **Predicted Positive** | **Predicted Negative** |
| ------------------- | ---------------------- | ---------------------- |
| **Actual Positive** | TP (True Positive ‚úÖ)   | FN (False Negative ‚ùå)  |
| **Actual Negative** | FP (False Positive ‚ùå)  | TN (True Negative ‚úÖ)   |

üìå **Meaning:**

- **TP: correctly predicted positive**

- **TN: correctly predicted negative**

- **FP: predicted positive but actually negative**

- **FN: predicted negative but actually positive**

```lua
    Confusion Matrix = [[TP, FP],
                        [FN, TN]]

```


## <center>**üîÑ Accuracy Score**</center>

> #### **Measures overall correctness of the model.**
$$Accuracy = \frac{TP + TN}{TP + TN + FP + FN}$$

‚úî **Good when dataset is balanced**

‚ùå **Misleading when dataset is imbalanced**


## <center>**‚úÖ Precision**</center>

> #### **Out of predictions that are positive, how many were correct?**

$$Precision = \frac{TP}{TP + FP}$$

##### **Useful When:**

- **False Positives are costly**

- **Example: Spam detection (don‚Äôt label real emails as spam!)**


## <center>**‚úÖ Recall (Sensitivity, True Positive Rate)**</center>

> #### **Out of actual positives, how many did we correctly predict?**

$$Recall = \frac{TP}{TP + FN}$$


Useful When:

- **False Negatives are costly**

- **Example: Cancer detection (never miss a positive patient!)**

## <center>**‚úÖ F1-Score**</center>

> #### **Balance between Precision and Recall**

$$F1 = 2 \times \frac{Precision \cdot Recall}{Precision + Recall}$$

‚úî Best when classes are imbalanced<br>
‚úî Helps avoid misleading high precision or recall alone<br>
‚úî Its is also helpful when we don't have enough domain knowledge to determine whether the evaluation of the model should be focused on Recall or Precision 

## <center>**ü¶∏ Support**</center>

> #### **Number of actual samples present for each class**

Example : 

```lua
    Support for class 1 = count of actual class 1 samples

```
- **Used to understand whether the dataset distribution is balanced**

## <center>**üìå ROC Curve & AUC Score**</center>


### <center>**üìå ROC ‚Äî Receiver Operating Characteristics**</center>

#### ‚úÖ **Origin & Meaning**

- **ROC** was first developed during **World War II**

- Goal: Evaluate **radar operators**

    - Radar system detects aircraft

    - Sometimes it wrongly detects birds/clouds as aircraft

So operators had **two possible mistakes**:

| Situation   | Meaning                                                | Example in ML                  |
| ----------- | ------------------------------------------------------ | ------------------------------ |
| False Alarm | Saying ‚Äúenemy aircraft detected‚Äù when actually nothing | Predicting cancer when healthy |
| Miss        | Failing to detect real aircraft                        | Predicting healthy when cancer |

Thus humans learned:

> ##### **We must measure not just correctness but also the types of errors we make.** 

> ##### That measurement became:
>‚úÖ ***ROC Characteristics***

<br>

#### **‚úÖ Why ‚ÄúReceiver Operating Characteristic‚Äù?**

| Word           | Meaning                                         |
| -------------- | ----------------------------------------------- |
| Receiver       | The radar **receiver** (like a classifier)      |
| Operating      | How it **operates** under different thresholds  |
| Characteristic | Its measured **ability** to distinguish signals |

üìå **Today, the radar receiver = machine learning classifier**<br>
üìå **Signal = Positive class**<br>
üìå **Noise = Negative class**<br>

**‚û° ROC tells how well a classifier separates Positive vs Negative**

<br>

#### **‚úÖ Connection With Modern ML**
- **A classifier outputs probability or score:**
$$P(y=1‚à£X)$$

But to decide a label, we select a **threshold**:

‚â• threshold ‚Üí Positive

< threshold ‚Üí Negative

Common default threshold = **0.5**
But in medical diagnosis, spam detection etc., **we should adjust the threshold**.

> ##### üëâ ROC shows what happens when we **change the threshold** from 1 ‚Üí 0

<br>

#### **‚úÖ What ROC Curve Plots**

| Axis   | Concept                                  | Formula                  |
| ------ | ---------------------------------------- | ------------------------ |
| Y-axis | True Positive Rate (Sensitivity, Recall) | $(TPR = \frac{TP}{TP+FN})$ |
| X-axis | False Positive Rate                      | $(FPR = \frac{FP}{FP+TN})$ |

<br>

> ##### **When threshold changes:**
> ##### **‚û° TP, FP, FN, TN change ‚Üí TPR & FPR change ‚Üí curve changes**


<br>


#### **‚úÖ Graphical Meaning**
| Point                                                       | Meaning                         |
| ----------------------------------------------------------- | ------------------------------- |
| (0,0)                                                       | Predict everything **Negative** |
| (1,1)                                                       | Predict everything **Positive** |
| Closer to (0,1) ‚Üí Ideal                                     |                                 |
| Diagonal Line ‚Üí Random Guessing                             |                                 |
| Below Diagonal ‚Üí Worse than random (reverse output to fix!) |                                 |

> ##### **üöÄ The more the curve bends toward top-left ‚Üí the better the model**


<br>


#### ‚úÖ **Why ROC is Powerful?**

- ‚úÖ **Does not depend on class imbalance**<br>
- ‚úÖ **Measures performance across all thresholds**<br>
- ‚úÖ **Helps compare different classifiers**<br>
- ‚úÖ **Shows error trade-off visually**<br>


<br>


---

### <center>**‚≠ê AUC ‚Äî Area Under Curve**</center>

> ##### ‚ÄúAUC‚Äù measures:

> ##### How much area lies under the ROC curve

**Range:**

$$0‚â§AUC‚â§1$$

| AUC           | Performance         |
| ------------- | ------------------- |
| **1.0** ‚úÖ     | Perfect model       |
| **0.90+**     | Excellent           |
| **0.80‚Äì0.90** | Good                |
| **0.70‚Äì0.80** | Fair                |
| **0.50** ‚ùå    | Random predictions  |
| **<0.50**     | Worse than guessing |


#### **‚úÖ Best Intuitive Definition of AUC**

| Threshold    | Decision    | Result                                             |
| ------------ | ----------- | -------------------------------------------------- |
| High (0.9)   | Very strict | Detect only clear positives ‚Üí Low FP, but Low TP   |
| Medium (0.5) | Standard    | Balanced FP & TP                                   |
| Low (0.1)    | Very loose  | Detect almost all positives ‚Üí High TP, but High FP |

üìå **ROC shows **this entire journey**, not just one point.**

## üìå When to use ROC-AUC?

Use when:

- **Classes are imbalanced**

- **Need probability-based comparison**

- **Want threshold-independent evaluation**

**Example scenarios:**

‚úÖ Cancer detection<br>
‚úÖ Fraud detection<br>
‚úÖ Spam filtering<br>


## ‚úÖ **ROC-AUC Summary**

| Concept   | Explanation                                | Formula        |
| --------- | ------------------------------------------ | -------------- |
| ROC Curve | TPR vs FPR at all thresholds               | ‚Äî              |
| TPR       | Detect true positives well                 | (TP/(TP+FN))   |
| FPR       | Mistakenly predicting positives            | (FP/(FP+TN))   |
| AUC       | Probability model correctly ranks examples | Area under ROC |


## üìï **Code Implementation**

````python
from sklearn.datasets import load_breast_cancer
import pandas as pd
import matplotlib.pyplot as plt
````

````python
data = load_breast_cancer()
````

````python
print(data.DESCR)
````

````python
data.target_names
````

````python
x = data.data
y = data.target
````

### üåüüåü **NOTE** :

> #### Actually in this preloaded breast_cancer_dataset `target value 0 represents` **Malignant** and `target value 1 represents` **Benign** which is very counter intuitive as we know that in logistic model and in other binary classfiers 1 is positive and 0 is negative and we should test for the patient being Malignant(tested positive)

> ***To achieve this we are going to use np.where(condition, if true value1, else value2)***

````python
import numpy as np
y = np.where(y == 0, 1, 0 )
````

````python
x.ndim, y.ndim
````

````python
# Now train test Split
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.20, random_state = 42) 
````

````python
# now importing logistic regression

from sklearn.linear_model import LogisticRegression
````

````python
lor = LogisticRegression(max_iter = 10000, C= 0.0002, penalty = 'l2')
````

````python
lor.fit(x_train, y_train)
````

````python
lor.score(x_test,y_test)
````

````python
y_pred = lor.predict(x_test)
y_pred_prob = lor.predict_proba(x_test)[:,1]
````

````python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report, roc_auc_score
````

````python
# 1Ô∏è‚É£ Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:\n", cm)

# 2Ô∏è‚É£ Accuracy
acc = accuracy_score(y_test, y_pred)
print("Accuracy:", acc)

# 3Ô∏è‚É£ Precision, Recall, F1-score, Support
print("\nClassification Report:\n")
print(classification_report(y_test, y_pred))

# 4Ô∏è‚É£ ROC-AUC Score
auc = roc_auc_score(y_test, y_pred_prob)
print("ROC-AUC Score:", auc)
````

````python
# Confusion matrix

from mlxtend.plotting import plot_confusion_matrix

plot_confusion_matrix(
    conf_mat= cm,
    colorbar= True,
    class_names= ["Benign(0)", "Malignant(1)"],
    
)
plt.xlabel("Predicted Values")
plt.ylabel("True Values")
plt.title("Confusion Matrix")
plt.show()
````

### ROC Curve Plot

````python
from sklearn.metrics import roc_curve
````

````python
fpr, tpr, thresholds = roc_curve(y_test, y_pred_prob)
````

````python
plt.figure()

plt.plot(fpr, tpr) #ROC curve
plt.plot([0,1], [0,1], linestyle = '--') # Diagonal reference line
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate (Recall)")
plt.title("Roc Curve")
plt.show()
````

````python
from mlxtend.plotting import plot_learning_curves

plot_learning_curves(x_train, y_train, x_test ,y_test, clf = lor)
plt.show()
````
